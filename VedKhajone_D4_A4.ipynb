**List**:
lists are ordered collections of items.Created using square brackets.Elements can be different data types.Lists are mutable,meaning you can modify their contents(add,remove,change elements).List items are ordered,changeable,and allow duplicate values.

**Example:**

```python
my_list =[1,2,3,'apple','banana']
my_list
```

**Tuple**

Tuples are similar to lists but are immutable(cannot be changed after creation)Created using parentheses.A tuple is a collection which is ordered and unchangeable.

**Example:**

```python
my_tuple =(1,2,3,'apple','banana')
my_tuple
```
